```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('max_columns', 50 )
```

```python
df = pd.read_excel('Sample - Superstore.xlsx')
df.head()
```

## Which Segment yielded the most profit

```python
plt.style.use('ggplot')
segment= df[['Segment', 'Profit']].groupby('Segment').sum('Profit')
plt.ylabel('Profit')
plt.bar(segment.index, segment['Profit'])
plt.title('Profit by Segment')
```

## Which Categories of product yielded the highest profit

```python
category= df[['Category', 'Profit']].groupby('Category').sum('Profit').sort_values('Category', ascending= False)
plt.bar(category.index, category['Profit'])
plt.ylabel('Profit')
plt.title(' Product Categories by Profit made')
```

## Which Sub-Categories yielded the highest profit

```python
plt.figure(figsize= (14,6))
sub_category= df[['Sub-Category', 'Profit']].groupby('Sub-Category').sum('Profit').sort_values(by= 'Profit', ascending= False)
plt.xticks(rotation=45)
plt.bar(sub_category.index, sub_category['Profit'])
plt.title('Profit by Sub-Categories')
#plt.text(500, 100, 's')
```

## Who are our most Frequent Customers

```python
customer= df[['Customer Name', 'Sales']].groupby('Customer Name').count()
customer =customer.sort_values(by = 'Sales', ascending= False).head(10).sort_values('Sales', ascending = 'True')
plt.barh(customer.index, customer['Sales'])
plt.title('Top Customers by No of purchases made')
plt.show()
```

## which states are the most and least profits made

```python
plt.figure(figsize= (14,6))
plt.subplot(1,2,1)
sub_category_top= df[['State', 'Profit']].groupby('State').sum('Profit').sort_values(by= 'Profit', ascending= False).head(10)
plt.xticks(rotation=45)
plt.bar(sub_category_top.index, sub_category_top['Profit'])
plt.title('Profit by Top State')
plt.subplot(1,2,2)
sub_category_bottom= df[['State', 'Profit']].groupby('State').sum('Profit').sort_values(by= 'Profit', ascending= True).head(10).sort_values('Profit', ascending=False)
plt.xticks(rotation=45)
plt.bar(sub_category_bottom.index, sub_category_bottom['Profit'])
plt.title('Profit by Bottom State')
plt.show()
#plt.text(500, 100, 's')
```
